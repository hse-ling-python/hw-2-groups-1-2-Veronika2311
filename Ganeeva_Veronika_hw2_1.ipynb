## Pymorphy

Токенизировать текст с помощью nltk. Эта ячейка может сообщить, что nltk уже установлен, но всё же

```python
import nltk
nltk.download('punkt')
```

```python
%%time

from nltk.tokenize import word_tokenize

tokens = word_tokenize(text)
print(tokens[0:100])
```

разобрать слова с помощью pymorphy, замерить время работы

```python
%%time
from pymorphy2 import MorphAnalyzer
morph = MorphAnalyzer()

result = {}

for number, token in enumerate(tokens):
    words = morph.parse(token)
    form = words[0]
    this_word = {}
    this_word['слово'] = form.word
    this_word['разбор'] = str(form.tag)
    this_word['лемма'] = form.normal_form
    this_word['вероятность'] = form.score
    result[number] = this_word

result
```

Сохраните результат (хотя бы лемма + часть речи) в json

```python
%%time
import json

with open('j_lemmas_pymprth.txt', 'w', encoding='utf-8') as f:
    json.dump(result, f, ensure_ascii = False)
```

Какую долю слов составляет каждая часть речи? 

Использовалась выдача pymorphy, программа выводит статистику по всему, что там выделено в отдельный класс

```python
parts = {'VERB':0, 'NOUN':0, 'ADJF':0, 'ADJS':0, 'COMP':0, 'INFN':0, 'PRTF':0, 
         'PRTS':0, 'GRND':0, 'NUMR':0, 'ADVB':0, 'NPRO':0, 'PRED':0, 'PREP':0,
         'CONJ':0, 'PRCL':0, 'INTJ':0}


for number, params in result.items():
    for p in params:
        if p == 'разбор':
            for part, co in parts.items():
                if params['разбор'].startswith(part):
                    new_co = co + 1
                    parts[part] = new_co

percents = {}
for part, co in parts.items():
    pcent = co/((len(result))/100)
    percents[part] = pcent
                    
                    
print(percents)
print(len(result))
```

Процент глаголов - в предыдущем задании.Топ-20 глаголов и наречий. Если список меньше 20ти, то он будет распечатан полностью

```python
import collections

all_verbs = []
all_advbs = []
for number, params in result.items():
    for p in params:
        if params['разбор'].startswith('VERB'):
            all_verbs.append(params['лемма'])
            
for number, params in result.items():
    for p in params:
        if params['разбор'].startswith('ADVB'):
            all_advbs.append(params['лемма'])
            

counter_verbs = dict(collections.Counter(all_verbs))
top_verbs = []
for key in sorted(counter_verbs, key=counter_verbs.get, reverse=True):
    top_verbs.append((key, counter_verbs[key]))

if len(top_verbs) > 20:
    print(top_verbs[0:20])            
else:
    print(top_verbs)

counter_advbs = dict(collections.Counter(all_advbs))
top_advbs = []
for key in sorted(counter_advbs, key=counter_advbs.get, reverse=True):
    top_advbs.append((key, counter_advbs[key]))

if len(top_advbs) > 20:
    print(top_advbs[0:20])            
else:
    print(top_advbs)
```

Биграммы nltk

Всё получилось так, потому что знаки препинания частотнее слов.

```python
from nltk.util import bigrams
from string import punctuation as punct

clear_tokens = []
for token in tokens:
    if not token in punct and not token in '--?!':
        clear_tokens.append(token)

top_bigrams = []
counter_bigrams = dict(collections.Counter(list(bigrams(clear_tokens))))
for key in sorted(counter_bigrams, key=counter_bigrams.get, reverse=True):
    top_bigrams.append((key, counter_bigrams[key]))
print(top_bigrams)
```
